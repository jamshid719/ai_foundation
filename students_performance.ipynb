```python
import pandas as pd
import numpy as np
```

```python
# NumPy vs Pandas
# Pandas = jadval bilan ishlash
# NumPy  = sonli array/matrix bilan tez ishlash

# | Pandas                              | NumPy                         |
# | ----------------------------------- | ----------------------------- |
# | DataFrame, jadval                   | Array, matrix                 |
# | Ustun nomlari bor                   | Ustun nomlari yo‘q            |
# | Data analysis uchun qulay           | Matematik hisoblash uchun tez |
# | `.groupby()`, `.value_counts()` bor | Matrix hisoblash kuchli       |
# | CSV/Excel bilan qulay               | ML model input uchun qulay    |
```

```python
df = pd.read_csv("StudentsPerformance.csv")

print("Dataset shape (rows, columns):", df.shape)
print("Column name:", df.columns.tolist())

print("First 5 rows of the dataset:")
print(df.head())

print("Summary statistics:")
print(df.describe())

print("Data types and null info:")
print(df.info())
```

```python
print("Average scores by subject:")
df[['math score', 'reading score', 'writing score']].mean()
```

```python
print("Max scores by subject:")
df[['math score', 'reading score', 'writing score']].max()
```

```python
print("Min scores by subject:")
df[['math score', 'reading score', 'writing score']].min()
```

```python
# Correlation - 2 uzgaruvchi bir-biri bn qanday bogliqligini kursatadigan kursatkich,
                #yani Bir qiymat uzgarganda, ikkinchisi qanchalik uzgarishi.
                # Corr. qanchalik 1 ga yaqin bulsa, bogliqlik shunchalik kuchli.
print("Correlation:")
df[['math score', 'reading score', 'writing score']].corr()
```

```python
df["total score"] = df[["math score", "reading score", "writing score"]].sum(axis=1)#axis=1 → qator (row) buyicha yigadi.
df["average score"] = df["total score"] / 3
df["passed"] = np.where(df["average score"] >= 50, "Yes", "No")# np.where → if/else kabi ishlaydi

print("New columns preview (total, average, passed):")
print(df[["total score", "average score", "passed"]].head())
```

```python
# Har bir col. un urtacha qiymatlarni hisoblaw

# groupby 'test preparation course' un 3 ta col. hisoblash(urtacha qiymatini)
test_prep_group = df.groupby('test preparation course')[["math score", "reading score", "writing score"]].mean()

print("Student score by test preparation course:")
print(test_prep_group)

# Gender-based average un 3 ta col. hisoblash(urtacha qiymatini)
gender_avg = df.groupby('gender')[["math score", "reading score", "writing score"]].mean()

print("Average score by gender:")
print(gender_avg)

# Who passed vs failed
pass_rate = df['passed'].value_counts(normalize=True)*100 # normalize=True - foizda qilib beradi
print("Pass rate:\n", pass_rate)
```

```python
# Numpy buyicha endi stat. amallarni bajarsak

#Oldin Pandas DataFrame’dan NumPy array kurinishiga utkazamiz.

#Convert scores to Numpy array
scores = df[["math score", "reading score", "writing score"]].values

# Mean, Std Deviation
print("Numpy Mean:", np.mean(scores, axis=0))
print("Numpy Std:", np.std(scores, axis=0))

# Z-score (normalize)
z_score = (scores - np.mean(scores, axis=0)) / np.std(scores, axis=0)
print("Z-scores (normalized numbers)")
print(z_score[:5])

# Average 0
# Negative -
# Positive +
```

```python
# How many students scored above 90 average?
top_scorers = df[df["average score"] >= 90]
print("Top scored student above 90:", len(top_scorers))

# How many failed?
failed = df[df["passed"] == "No"] # passed ustunida "No" bulganlarni oliw
print("Failed from exam:", len(failed))
```

```python
import numpy as np

arr = np.array([1,3,5,7])
print('numpy_array:', arr)
```

```python
arr2 = np.array([[1, 3, 5, 7], [4, 6, 8, 9]])
print('numpy_array2:', arr2)


arr2.shape #arr shape
```

```python
# dtype()
a = np.array([1, 2, 3])
print(a.dtype)
```

```python
# reshape() - arr. shapeni uzgartirish
a = np.array([1, 2, 3, 4, 5, 6])
b = a.reshape(2, 3)
print("shaped_array:", b)
```

```python
# astype() - arr. ni type ni uzgartirish
a = np.array([1, 2, 3])
b = a.astype(float)
print("float:", b)
```

```python
# np.zeros(), np.ones() - 0 va 1 dan tashkil topadigan arr yaratish => dummy data
z = np.zeros((2, 3))
o = np.ones((2, 3))

print("zeros:", z)
print("ones:", o)
```

```python

# ndim - deminsionni aniqlash (nechta vektorligini bilish un)
a = np.array([[1, 2], [3, 4]])
print("ndim:", a.ndim)
```

```python
# arange() - berilgan numberlar orqali arr. yaratadi "[ )" (bu test arr. yaratiw un kk)
a = np.arange(1, 6)
print("arange:", a)
```

```python
# sum(), mean(), max(), min()
a = np.array([1, 2, 3, 4])

print("sum:", a.sum())
print("mean:", a.mean())
print("max:", a.max())
print("min:", a.min())
```

```python
# flatten() - 2d, 3d arr.larni 1d arr. list ga aylantiradi(rasmlar bn ishlashda qullaniladi)
a = np.array([[1, 2], [3, 4]])
print("flatten:", a.flatten())
```

```python
# T (Transpose) method - bu arr.lar ichidagi num.larni random sifatida almashtiradi yoki
a = np.array([[1, 2], [3, 4]])
print("transpose:", a.T)
```

```python
# dot() @ - Matrix multiplication da a ning qatori bilan b ning ustuni kupaytirilib qushiladi.
a = np.array([[1, 2], [3, 4]])
b = np.array([[2, 0], [1, 2]])
print("dot/multiplication:", a.dot(b))
```

```python
# random.rand() - 0 va 1 oraligida random arr.lar yaratadi.
a = np.random.rand(2, 3)  # => 0.0 and 1.0 / 0.0 included / 1.0 excluded
print("random rand:", a)
```

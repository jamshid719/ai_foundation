```python

data_with_missing = {
    "cars": ["BMW", "Volvo", None, "Tesla", "Ferrari", None, "Ford"],
    "counts": [3, 7, None, 8, 1, 9, None]
}

df_missing = pd.DataFrame(data_with_missing)

print(df_missing)
```

```python
#datalarni cleaning qilish methodi(tekshiriw)

df_missing.isnull().sum() # mavjud bulmagan datalarni xisoblaydi.
                          # df_missing.isnull() bulsa, yuq datalarni true qb, borlarini false qilib chiqaradi
```

```python
df_missing.dropna() # faqat mavjud datalarni olib beradi
```

```python

df_missing.dropna(how="all") # 3 qatorni olib tashlaydi, chunki shu qatorda ham "cars" da ham "count" data yuq
```

```python
df_missing.dropna(axis=1) # data yuq bulgan columnlarni butunlay uchirib tashlaydi
```

```python
df_missing.dropna(subset=['counts']) # faqat countsda yuq datalarni olib tashlaydi, "cars" columni tekshirmaydi
```

```python
df_missing.fillna(0) # mavjud bulmagan kataklarga 0 quyib chiqadi.(0 urniga harqanday value quyish mumkin, kop holatlarda "unknown" ham quyib ketadi)
```

```python
df_missing.fillna({"cars": "unknown", "counts": 0}) # cars ga "unknown",  "counts"ga 0 quyadi
```
